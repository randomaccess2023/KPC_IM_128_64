```python
# import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./64/models14/hist_modelS_3000', sep='\t', index_col=False) # load history
```

```python
lambda_marginal_entropy = 0.1
lambda_conditional_entropy = 0.03
```

```python
# plot the learning curve

plt.figure(figsize=(8, 5))

plt.plot(df['tt'], df['loss'], 'r', label='Loss function', linewidth=3)
plt.plot(df['tt'], df['loss_rec'], 'g', label='Reconstruction loss', linewidth=3)
plt.plot(df['tt'], lambda_marginal_entropy * np.array(df['entropy_marginal']) - lambda_conditional_entropy * \
         np.array(df['entropy_mean']), 'b', label='Mutual information (MI)', linewidth=3)

plt.xlabel('Number of epochs', fontsize=22)
plt.ylabel('Value', fontsize=22)
plt.title('The learning curve', fontsize=30)
plt.legend(loc='best', fontsize=15, bbox_to_anchor=(0.03, 0.15, 0.755, 1.27));
```
